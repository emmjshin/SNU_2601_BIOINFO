```python
import pandas as pd

pileup = pd.read_csv('CLIP-let7g-gene.pileup', sep='\t', names=['chrom', 'pos', '_ref', 'count', 'basereads', 'quals'])
pileup.tail()
```

```python
import re
toremove = re.compile('[<>$*#^]')
pileup['matches'] = pileup['basereads'].apply(lambda x: toremove.sub('', x))
```

```python
%%R -i pileup
library(tidyverse)

# Function to calculate entropy
calc_entropy <- function(s) {
  if (s == "") return(0)
  chars <- strsplit(toupper(s), "")[[1]]
  probs <- table(chars) / length(chars)
  -sum(probs * log2(probs))
}

# 1 & 2. Calculate entropy
result <- pileup %>%
  rowwise() %>%
  mutate(entropy = calc_entropy(matches)) %>%
  ungroup()

# 3. Format as bedGraph
bedgraph_r <- result %>%
  mutate(start = pos - 1, end = pos) %>%
  select(chrom, start, end, entropy)

print(head(bedgraph_r))

# 4. Save for Google Drive
write.table(bedgraph_r, "entropy_r.bedgraph", sep="\t", quote=F, row.names=F, col.names=F)
```
